```python
# 다중 변수 동시 선언
x, y, z = "Apple", "Banana", "Carrot"
print(x, y, z)

# 밑줄(_)로 변수명을 지정하는 경우는 사용하지 않을 변수를 위한 자리 표시 용도로 활용.
_, var = "Not use", "Use"
print(_, var)
```

```python
# 단일 값의 다중 변수 할당
# 단일 값의 다중 변수는 쉼표가 아닌 등호로 연결하기.
x = y = z = "Dog"
print(x, y, z)
```
