```python
# 리스트 생성하기
fruit_list = ["apple", "banana", "orange"]
print(fruit_list)
print(type(fruit_list))
print('\n')

# 리스트는 중복 허용.
# 순서대로 저장함.
fruit_bucket = ["apple", "banana", "orange", "apple", "banana"]
print(fruit_bucket)
print(type(fruit_bucket))
```

```python
# 인덱스 사용하기
# 인덱스를 이용해 슬라이싱도 가능.
fruit_list = ["apple", "banana", "orange"]
print(fruit_list[0])
print(fruit_list[-1])
print(fruit_list[1:3])
```

```python
# 리스트는 가변(mutable) 자료형.
# 리스트 생성 이후에도 아이템 값을 변경 가능.
fruit_list[1] = "kiwi"
print(fruit_list)
```

```python
# 리스트에 새 요소 추가하기.
# insert() 함수: 지정한 인덱스 위치에 요소를 추가.
# append() 함수: 리스트의 마지막 위치에 요소를 추가.
# extend() 함수: 다른 리스트의 요소들을 결합.
# +: 리스트를 합쳐 하나의 새 리스트를 생성.
fruit_list.insert(2, "mango")                       # 인덱스 값 2의 위치에 "mango" 요소를 추가.
print(fruit_list, '\n')

fruit_list.append("watermelon")                     # 마지막 위치에 "watermelon" 요소 추가.
print(fruit_list, '\n')

vegitable_list = ["carrot", "tomato", "onion"]
fruit_list.extend(vegitable_list)
print(fruit_list, '\n')

list1 = [1, 2, 3]
list2 = ['가', '나', '다']
list3 = list1 + list2
print(list3)                                        # '+' 기호로 리스트 두 개를 결합해 하나의 새 리스트 생성.
```

```python
# 리스트의 요소 삭제
# remove() 함수: 지정한 요소를 삭제.
# del: 특정 위치의 지정 요소 또는 리스트 전체를 삭제.
# clear(): 리스트 내의 모든 요소를 삭제.
fruit_list = ["apple", "tomato", "banana", "orange"]
fruit_list.remove("tomato")
print(fruit_list, '\n')

del fruit_list[-1]
print(fruit_list, '\n')

'''del fruit_list              # fruit_list 라는 리스트 자체를 삭제.
print(fruit_list, '\n')     # "NameError: name 'fruit_list' is not defined." 에러 발생.'''

fruit_list = ["apple", "banana", "orange"]
fruit_list.clear()
print(fruit_list)
```

```python
# 리스트 내부 정렬
# sort() 함수: 리스트 내부의 요소들을 오름차순 또는 내림차순으로 정렬.
# 'reverse=True' 옵션 명시 시 내림차순 정렬, 옵션을 명시하지 않으면 기본값으로 오름차순 정렬.
fruit_list = ['strawberry', 'mango', 'blueberry', 'watermelon', 'apple', 'banana', 'orange']
fruit_list.sort()           # 오름차순 정렬
print(fruit_list, '\n')

fruit_list.sort(reverse=True)       # 내림차순 정렬
print(fruit_list)
```
