```python
# 튜플(tuple)
# 리스트와 달리 '()'로 생성하고, 내부는 쉼표로 구분.
# 각 요소가 순서를 가짐.
# 리스트와 달리 요소의 변경, 추가, 삭제가 불가능한 불변 속성을 지님.
fruit_tuple = ("apple", "banana", "orange")
print(fruit_tuple, type(fruit_tuple), '\n')

# 튜플도 리스트처럼 중복된 요소 저장 가능.
fruit_tuple = ("apple", "banana", "orange", "apple", "banana")
print(fruit_tuple)
```

```python
# 튜플도 인덱스 사용 가능.
fruit_tuple = ("apple", "banana", "orange")
print(fruit_tuple[1], '\n')

'''# 튜플은 불변 자료형이기 때문에, 요소 값 변경(수정)이 불가능함.
fruit_tuple[1] = "kiwi"         # 실행 시 "TypeError: 'tuple' object does not support item assignment." 에러 발생.
print(fruit_tuple, '\n')'''

'''# 요소 추가 및 삭제도 불가능함.
fruit_tuple.append("watermelon")        # 실행 시 "AttributeError: 'tuple' object has no attribute 'append'" 에러 발생.
print(fruit_tuple)
fruit_tuple.remove("apple")             # 실행 시 "AttributeError: 'tuple' object has no attribute 'remove'" 에러 발생.
print(fruit_tuple)'''

```

```python
# 튜플 내의 요소를 수정하려면, 리스트로 변환 후 요소를 변경하는 우회 방법을 사용해야 함.
fruit_list = list(fruit_tuple)      # 튜플을 list()로 감싸서 리스트로 변환. 이후에 list에서 사용하는 함수들을 사용해서 요소 작업.
fruit_list.append("watermelon")
print(fruit_list, '\n')

fruit_list.remove("apple")
print(fruit_list, '\n')

fruit_tuple = tuple(fruit_list)     # 수정 작업을 마쳤으면 다시 tuple로 변환하기.
print(fruit_tuple)
```

```python
# 세트(set)
# 중괄호{}를 사용.
# 집합의 성격을 가지며, 순서를 갖지 않고 중복을 허용하지 않음.
# 따라서 인덱스도 사용 불가능.
fruit_set = {"apple", "banana", "orange"}
print(fruit_set, type(fruit_set), '\n')

# 중복된 요소가 포함되게 세트를 생성해도, 실제로는 중복되게 저장되지 않음.
fruit_set = {"apple", "banana", "orange", "apple", "banana"}
print(fruit_set)
```

```python
# 세트는 순서가 없는 자료형이기 때문에, 인덱스 사용 불가능. 인덱스 사용 시 에러 발생.
print(fruit_set[1])        # 실행 시 "TypeError: 'set' object is not subscriptable." 에러 발생.
```

```python
# 단, 세트에 요소 추가 가능.
# 다른 세트의 요소도 추가 가능.
# add() 함수: 단일 요소 추가.
# update() 함수: 다른 세트의 요소를 추가.
fruit_set.add("kiwi")
print(fruit_set, '\n')

vegetable_set = {"carrot", "tomato", "onion"}
fruit_set.update(vegetable_set)
print(fruit_set)
```

```python
# 세트의 요소 삭제
# remove() 함수: 지정한 요소를 삭제.
# del: 세트 자체를 삭제.
# clear() 함수: 세트의 모든 요소들을 삭제.
fruit_set.remove("onion")
print(fruit_set)
```
