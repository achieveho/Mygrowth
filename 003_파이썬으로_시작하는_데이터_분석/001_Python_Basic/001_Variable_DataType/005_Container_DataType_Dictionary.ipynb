```python
# 딕셔너리(dictionary)
# 중괄호 사용
# '키(Key):값(Value)' 쌍으로 구성되어 있음.
my_dict = {
    "name": "Harry",
    "age": 27,
    "height": 190,
    "weight": 99.9
}
print(my_dict, type(my_dict))
```

```python
# 요소 선택, 추가
# .keys() 메서드를 사용하면 모든 키 확인 가능.
print(my_dict.keys(), '\n')

# 하나의 key만 지정하면, 그 key의 value 확인 가능.
print(my_dict["age"], '\n')

# key 값을 지정하여 그 key의 value 값 수정 가능.
my_dict["age"] = 25
print(my_dict, '\n')

# update() 메서드를 사용해 기존 '키-값' 쌍을 업그레이드 하거나 새 '키-값'쌍 추가 가능.
my_dict.update({"weight": 100})
print(my_dict, '\n')

my_dict.update({"address": "Busan"})
print(my_dict)
```

```python
# 딕셔너리의 요소 삭제
# 리스트와 다름.
# popitem() 메서드: 마지막 한 쌍을 삭제.
# pop() 메서드: 특정 위치의 한 쌍을 삭제.
# clear() 메서드: 모든 쌍을 삭제.
my_dict.popitem()       # 마지막 한 쌍을 삭제.
print(my_dict, '\n')

my_dict.pop("age")      # age라는 key 삭제.
print(my_dict, '\n')

my_dict.clear()         # 딕셔너리가 삭제되는 것이 아닌, 형태는 유지된 채 내부만 삭제됨.
print(my_dict)
```
