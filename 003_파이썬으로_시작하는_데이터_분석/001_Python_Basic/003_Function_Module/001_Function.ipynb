```python
# 함수 사용의 장점
# 01. 코드 재사용성: 동일한 기능을 여러 번 사용 가능.
# 02. 가독성 향상: 특정 기능을 함수로 묶어 코드를 체계적으로 관리.
# 03. 유지보수 용이: 기능별로 분리되어 수정과 디버깅이 쉬움.
# 문제 01. 기온이 0도보다 높으면 '아이스 아메리카노'를 출력, 그렇지 않은 경우 '따뜻한 아메리카노'를 출력하는 함수 작성하기.
def americano(temp):
    if temp > 0:
        print("아이스 아메리카노")
    else:
        print("따뜻한 아메리카노")

temp = int(input("현재 기온(숫자로만 입력): "))
americano(temp)
```

```python
# 문제 01-1. 함수 안에서 return 사용하기.
def americano(temp):
    if temp > 0:
        result = "아이스 아메리카노"
    else:
        result = "따뜻한 아메리카노"
    
    return result

temp = int(input("현재 기온을 숫자로만 입력: "))
print(f"현재 기온은 {temp}도 이므로, {americano(temp)}를 추천드립니다!")
```

```python
# 문제 02. 1학년 2반의 시험 점수는 [80, 90, 70, 65, 85, 95, 90, 85, 75, 80] 이다.
# 한 문제가 잘못되어 모두 5점의 추가 점수를 받아야 할 때, 추가 점수가 적용된 점수를 반환하는 함수 작성하기.
def score_updater(scores):
    for i in scores:
        i += 5
        up_scores.append(i)
    
    return up_scores

scores = [80, 90, 70, 65, 85, 95, 90, 85, 75, 80]
up_scores = []

print(f"1학년 2반의 최종 점수: {score_updater(scores)}")
```

```python
# 문제 03. 산토끼 동요 노래에서 '토'가 나오는 횟수를 구하는 함수 작성.
def to_count(song):
    num = 0
    for i in song:
        if i == '토':
            num += 1
        else:
            pass
    
    return num

song = '산토끼 토끼야. 어디를 가느냐. 깡총깡총 뛰면서. 어디를 가느냐.' \
'산고개 고개를. 나혼자 넘어서. 토실토실 알밤을. 주워 올테야.'
# num = 0

print(f"해당 노래에서 '토'는 총 {to_count(song)}개 나옵니다.")
```

```python
# 문제 03에 대한 교재의 답안
def get_char_count(lylic, char):
    count = 0
    for i in lylic:
        if i == char:
            count += 1
    
    return count

lylic = """산토끼 토끼야. 어디를 가느냐. 깡총깡총 뛰면서. 어디를 가느냐.
산고개 고개를. 나혼자 넘어서. 토실토실 알밤을. 주워 올 테야."""

print(get_char_count(lylic, '토'))
```

```python
# 문제 04. 영단어 또는 문장을 인자로 받아 대문자와 소문자로 출력하는 함수 작성하기.
def change_up_low(eng):
    up_result = eng.upper()
    lo_result = eng.lower()

    return up_result, lo_result

eng = input("영단어 또는 영어로 된 문장을 입력하세요: ")

up_result, lo_result = change_up_low(eng)                  # '튜플 언패킹'을 사용해 여러 개의 함수의 반환값을 각각의 변수에 저장하여 편하게 사용.

print(f"입력한 글자를 모두 대문자로 변환한 결과: {up_result}")
print(f"입력한 글자를 모두 소문자로 변환한 결과: {lo_result}")
```

```python
# 문제 05. 두 수와 연산자를 입력받아 사칙연산 결과를 출력하는 계산기 함수 작성하기.
def calculator(num1, num2, operator):
    result = 0
    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 == 0:
            print("0으로는 나눌 수 없습니다.")
            return
        result = num1 / num2
    else:
        print("올바르지 않은 연산자를 입력하여, 연산을 진행할 수 없습니다.")
    
    return print(f"{num1} {operator} {num2} = {result}")

num1, num2 = map(int, input("두 정수를 입력하세요: ").split())
operator = input("'+', '-', '*', '/' 중 하나의 연산자를 입력하세요: ")

calculator(num1, num2, operator)
```

```python
# 문제 06. 사용자로부터 키와 성별을 입력 받아 권장 체중을 출력하는 함수 작성하기.
# 남성 권장 체중: (키 - 100) * 0.9
# 여성 권장 체중: (키 - 100)
def recommend_weight(height, sex):
    result = 0

    if sex == '남' or '남성' or '남자':
        result = (height - 100) * 0.9
        return print(f"사용자님의 권장 체중은 {result}kg 입니다.")
    elif sex == '여' or '여자' or '여성':
        result = (height - 100)
        return print(f"사용자님의 권장 체중은 {result}kg 입니다.")
    else:
        print("잘못된 성별을 입력하셨습니다.")
        return

height = float(input("키를 cm를 제외한 소수점 첫째 자리로 입력하세요: "))
sex = input("성별을 입력하세요: ")

recommend_weight(height, sex)
```

```python
# 문제 06의 교재 답안.
# default 인자 사용
# defalut 인자는 반드시 일반 인자 뒤에 위치해야 함.
def print_weight(height, man=True):
    weight = 0
    if man:
        weight = (height - 100) * 0.9
    else:
        weight = (height - 100)
    
    print("권장 체중은 {}kg 입니다.".format(weight))

print_weight(170)           # 디폴트값을 사용해 성별을 별도로 입력하지 않고도 남성 몸무게를 계산.
print_weight(170, True)     # 명시적으로 man이 True라는 것을 전달.
print_weight(170, False)    # 명시적으로 man이 False임을 전달해 여성의 권장 체중 공식을 사용.
```

```python
# 문제 07. *args(가변 인자)를 사용해 입력 받은 인자를 모두 더하는 함수 작성하기.
def sum_machine(*args):
    result = 0
    for num in args:
        result += num
    
    return print(f"입력한 수를 모두 합한 값: {result}")

material = map(int, input("덧셈을 원하는 정수들을 입력하세요(공백으로 구분): ").split())
sum_machine(*material)
```

```python
# 문제 08. 섭씨 온도를 화씨 온도로 변환하는 프로그램 함수 작성하기.
# 섭씨 -> 화씨: (섭씨 * 9/5) + 32
# 화씨 온도를 소수점 첫째 자리까지만 출력.
def celsius_to_fahrenheit(celsius):
    fahrenheit = float((celsius * 9/5) + 32)

    return print(f"섭씨 {celsius}도는 화씨로 {fahrenheit:.1f}도 입니다.")

temperature = int(input("현재 기온을 정수로 입력하세요: "))
celsius_to_fahrenheit(temperature)
```
