```python
# 입출력(Input, Output)
# input() 함수로 사용자에게 데이터를 입력받음.
# output() 함수로 결과를 출력.
```

```python
# 입력
# input() 함수
# 사용자로부터 키보드로 입력을 받을 때 사용하는 함수.
# 사용자가 입력한 내용을 문자열로 반환.
name = input("이름을 입력하세요: ")
print(name + "님, 안녕하세요!")
```

```python
# input() 함수는 항상 기본적으로 문자열로 데이터를 반환함.
# 다른 자료형을 원할 경우 데이터 타입을 변환해야 함.
# 계산 결과를 다른 문자열과 연결해 출력할 때는 str() 함수를 사용해 숫자를 문자열로 변환해야 함.
height = input("키를 입력하세요: ")
weight = (float(height) - 100) * 0.9
print("권장 체중은 " + str(weight) + "kg 입니다.")
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# 다양한 입출력 사례
# print(): 기본 출력
print("apple")

# print(a, b, c): 여러 값 연결 출력
print("apple", "banana", "carrot")

# print(sep=): 구분자 지정
print("a", "b", sep=",")

# print(end=): 줄바꿈 제어
print("text", end="")

# "{}".format(): 포맷 스트링
print("{}을 먹고싶다.".format("치킨"))

# {:.3f}: 소수점 자리수 지정
"{:.3f}".format(3.14159)

# {:,}: 천 단위 콤마
"{:,}".format(1000000)
```

```python
# print() 함수는 여러 개의 값을 동시에 출력 가능.
# 기본적으로 여러 값들 사이에는 공백이 자동으로 삽입되어 구분됨.
print("apple", "peach", "mango")
```

```python
# sep 매개변수 - 구분자 변경
# 여러 값들 사이의 구분자를 원하는 문자로 변경 가능.
# 기본값은 공백, 이를 쉼표, 하이픈 또는 다른 문자로 변환 가능.
print("apple", "peach", "mango", sep=",")
```

```python
# end 매개변수 - 줄바꿈 제어
# print() 함수는 기본값으로 출력 후 자동 줄바꿈을 실행함.
# end 매개변수 사용 시 이 동작을 제어 가능.
# 예를 들어 end=""로 설정 시 줄바꿈 없이 다음 출력이 같은 줄에 이어짐.
print("원숭이 엉덩이는 빨개 ", end="")
print("빨가면 사과")
```

```python
# 개행문자 () 활용
# 대표적인 개행문자 \n
# 줄바꿈 수행.
# 여러 줄에 걸친 텍스트를 하나의 print() 함수로 출력 시 유용.
print("산토끼 토끼야\n어디를 가느냐\n깡총깡총 뛰면서\n어디를 가느냐")
```

```python
# 문자열 포맷팅 - format() 메서드
# 동일한 형태의 문자열에서 특정 부분만 변경해 출력하고자 할 때 사용.
# 템플릿 형태의 문자열에 변수 값을 쉽게 삽입할 수 있어 코드의 가독성 증가.
# 기본 포맷 스트링은 {}를 사용해 변수가 들어갈 위치를 표시하고 format() 메서드로 값을 전달.
food = "치킨"
print("나는 {}을 먹고 싶어요.".format(food))
```

```python
# format() - 다중 값 치환
# 여러 개의 중괄호를 사용해 한 번에 여러 값을 문자열에 삽입 가능.
food1 = "치킨"
food2 = "피자"
print("나는 {}과 {}를 먹고 싶어요.".format(food1, food2))
```

```python
# format() - 인덱스를 사용한 위치 지정
# 중괄호 안에 숫자를 넣어 format() 메서드의 인수 순서를 명시적으로 지정 가능.
# 같은 값을 여러 번 사용하거나 순서를 바꿔 사용 가능.
print("나는 {0}, {1}을 먹고 싶다. 우리 집엔 {1}이 배달되지 않아 슬프다.".format("피자", "치킨"))
```

```python
# 키워드를 활용한 명명
# 중괄호 안에 키워드 사용 시 format() 메서드에서 해당 키워드로 값을 전달함.
# 코드의 의미를 더욱 명확히 만들어 줌.
text = "{name}님, 반갑습니다. 적립금은 {money}원 입니다."
print(text.format(name = "홍길동", money = 500))
```

```python
# 숫자 포매팅
# 숫자의 표시 형식을 다양하게 제어 가능.

# 소수점 자릿수 지정
# : 뒤에 .@f를 사용하면 지정한 소수점 자릿 수 까지만 표현함.
print("{:.2f}% 확신합니다.".format(95.1234567))
```

```python
# 숫자 포매팅
# 천 단위 콤마 표시
# : 뒤에 쉼표 사용 시 큰 숫자에 천 단위 구분 기호가 자동으로 삽입됨.
print("한 달 휴대폰 요금은 {:,}원 입니다.".format(100000))
```
