```python
# 파일 입출력
# 파일을 다루는 과정 = '파일 입력 과정' + '파일 출력 과정'
# 파일 입력 과정: 파일을 쓰기 모드로 열고, 파일에 내용을 작성하고, 파일을 닫기. (Open, Write, Close)
# 파일 출력 과정: 파일을 읽기 모드로 열고, 파일의 내용을 읽어오고, 파일을 닫기. (Open, Read, Close)
```

```python
# 파일 입력
# 파일을 생성하고 내용을 작성하는 코드.
f = open("abc.txt", "w")
f.write("A B C D E F G ")
f.close()

# open() 함수의 옵션
# 'r': 읽기 모드 / 이미 있는 파일의 내용을 읽어올 때 사용.
# 'w': 쓰기 모드 / 파일에 새로운 내용 작성 시 사용, 기존 내용을 모두 지우고 덮어씀. / 해당 파일의 내용이 없으면 새 파일을 자동으로 생성함.
# 'a': 추가 모드 / 기존 파일의 내용을 그대로 두고 뒤에 이어서 새로운 내용을 덧붙일 때 사용.
```

```python
# 쓰기 모드 'w'
f = open("abc.txt", 'w')
f.write("a, b, c, d, e, f, g ")
f.close()

f = open("abc.txt", 'r')
print(f.read())
f.close()
```

```python
# 추가 모드 'a'
f = open("abc.txt", 'a')
f.write(" H, I, J, K, L, M, N, O, P, Q, R, S, T, U, V, W, X, Y, Z")
f.close()

f = open("abc.txt", 'r')
print(f.read())
f.close()
```

```python
# readlines(): 줄 단위로 읽기
# 파일을 줄 단위로 나눠 리스트 형태로 반환함.
# 여러 줄로 구성된 파일을 처리할 때 유용함.
f = open("abc.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)

f.close()
```

```python
# 파일 출력 - 파일의 내용 읽기
f = open("abc.txt", "r")        # 읽기모드로 파일 열기.
print(f.read())
f.close()
```

```python
# with 구문 활용하기
# close()를 잊어버린느 실수를 방지하기 위해 사용.
# 블록을 벗어날 때 자동으로 파일이 닫힘.

# 파일을 쓰기모드로 열어서 새로운 내용 작성.
with open("일기.txt", "w") as f:
    f.write("2025년 12월 17일 수요일 16:10\n")      # 날짜를 파일에 적고 줄바꿈 실행.

# 파일을 추가모드로 열어서 기존 내용 뒤 새 내용 추가.
with open("일기.txt", "a") as f:
    f.write("날씨: 맑음")                         # 기존 내용을 지우지 않고, 이어서 새 내용 추가.

# 파일을 읽기모드로 열어, 파일의 전체 내용 확인.
with open("일기.txt", "r") as f:
    print(f.read())                                 # 파일의 모든 내용을 화면에 출력.
```

```python
# 파일 삭제
# 파이썬에서 파일을 삭제할 때는 'os' 모듈을 사용해야 함.
# os 모듈은 윤영체제와 관련된 기능을 제공하는 파이썬의 내장 모듈임.
# 파일 삭제 전, 반드시 해당 파일의 존재 여부를 확인해야 함.
# 존재하지 않는 파일을 삭제하려 하면 에러 발생.

import os       # 운영체제 관련 기능을 사용하기 위해 os 모듈을 import.

file = "abc.txt"        # 삭제를 원하는 파일명을 변수에 저장.

#os.path.exists()        # 파일의 존재 여부를 확인하는 구문.
if os.path.exists(file):
    os.remove(file)     # 파일이 존재하면 'os.remove()' 메서드를 사용해 파일을 삭제.
    print(f"{file}파일을 정상적으로 삭제하였습니다.")
else:
    print(f"{file}파일이 존재하지 않습니다.")
```
