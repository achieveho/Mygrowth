```python
# 문자열
# 문자열 함수
# split(): 문자열을 분리.
# strip(): 문자열 삭제.
# join(): 문자열 연결.
# find(): 문자열 찾기.
# count(): 문자열 일치 횟수 반환.
# startswith(): 특정 문자열로 시작하는지 여부를 검사.
# endswith(): 특정 문자열로 끝나는지 여부를 검사.
# replace(): 문자열 바꾸기.
# lower(): 소문자로 변경.
# upper(): 대문자로 변경.
```

```python
# split() 함수
# 문자열을 분리하는 함수.
# 지정한 구분자로 나누고, 리스트로 반환함.
text = "나는 자랑스러운 태극기 앞에 자유롭고 정의로운 대한민국의 무궁한 영광을 위하여 충성을 다할 것을 굳게 다짐합니다."
sp_text = text.split(" ")     # 공백을 구분자로 지정해 문자를 분리하고, 리스트에 대입.

phone_number = "+82-10-1234-5678"
dash_pnum = phone_number.split("-")
twodash_pnum = phone_number.split("-", 2)       # 구분자 뒤 옵션은 최대로 분리할 횟수.

print(sp_text)
print(dash_pnum)
print(twodash_pnum)
```

```python
# strip() 함수
# 문자열의 양쪽 끝에서 지정한 문자를 제거.
# 기본값은 공백 문자 제거.
text01 = "    토실토실 아기 돼지    "
st_text01 = text01.strip()

text02 = "\n\n\n토실토실 아기 돼지\n\n\n"
st_text02 = text02.strip("\n")

text03 = "ababab토실토실 아기 돼지ababab"
st_text03 = text03.strip("ab")

text04 = "ssss토실토실 아기 돼지sssss"
lst_text04 = text04.lstrip("s")
rst_text04 = text04.rstrip("s")

print(st_text01)
print(st_text02)
print(st_text03)
print(lst_text04)
print(rst_text04)

# 만일 양쪽 끝의 지우려는 문자가 다른 경우, '.replace()'를 한 줄에 두 번 사용하는 것이 더 좋음.
text05 = "abcde스트립쓰지말기fghij"
rp_text05 = text05.replace("abcde", "").replace("fghij", "")
# print("\n",rp_text05) ==> 이렇게 쓰면 'sep=""' 때문에 왼쪽에 공백 한 칸이 같이 출력됨.
# 그래서 원하는 내용을 출력하기 전 줄바꿈을 먼저 하고 싶다면, f-string을 사용해야 함.
print(f"\n{rp_text05}")
```

```python
# join() 함수
# 반복 가능한 객체의 요소들을 문자열로 연결함.
text01 = '->'
text02 = ["인천", "도쿄", "뉴욕", "파리"]
jtext = text01.join(text02)
print(jtext)
```

```python
# 교재의 join() 예제가 너무 별로라 ChatGPT에게 예제 5개 생성을 부탁함.
# join()의 기본 형태: '구분자.join(반복가능객체)'
# join의 주체는 문자열임.
# 인자는 반복 가능한 객체(iterable)임. ==> list, tuple, set, dict, str
# 반복 가능한 객체의 모든 요소는 반드시 문자열이어야 함.

# 예제 01
sep = " -> "
cities = ["인천", "도쿄", "뉴욕", "파리"]

result = sep.join(cities)
print(result)

# 구분자는 각 요소의 사이에만 들어감.

# 예제 02
# .join()을 실행하기 위해 내부에서는 ["P", "y", "t", "h", "o", "n"]으로 처리됨.
# 그래서 각 문자 사이에 '-'가 삽입됨.
# 문자열을 문자 단위로 분해하여 재조합 할 때 유용함.
# '토큰화/포맷 변환'의 기초 개념.
text = "Python"
result = "-".join(text)
print(f"\n{result}")

# 예제 03
# 문자열이 아닌 다른 것을 join의 인자로 사용할 경우, TypeError(sequence item 0: expected str instance, int found)가 발생.
# 무조건 .join()의 인자는 문자열이어야 함.
# 아래는 에러가 발생하는 코드
# data = [2024, 2025, 2026]
# result = ",".join(data)
# print(data)

data = [2024, 2025, 2026]
result = ", ".join(map(str, data))
print(f"\n{result}")

# 예제 04. join을 문자열 간 단순 연결에도 사용 가능.
# 토큰화 된 문자 또는 단어를 재조합 할 때 사용.
# 전처리 후 문자열 복원 시 사용.
# NLP 전처리 단계에서 사용.
# 연산자의 반복보다 성능 및 가독성 모두 우수함.
chars = ["데", "이", "터", "분", "석"]
result = "".join(chars)
print(f"\n{result}")

# 예제 05
# 경로, 로그, 식별자 생성
# 장점 01. 요소 수가 늘어나도 코드의 변경 불필요.
# 장점 02. 반복문 없이 깔끔함.
# 장점 03. 데이터 기반 문자열 생성에 최적화.
path_parts = ["home", "achieveho", "projects", "python"]
path = "/".join(path_parts)
print(f"\n{path}")

# join()은 문자열 포맷 함수가 아니라, '문자열 사이에 규칙적으로 끼워 넣는 결합 연산'임.
```

```python
# find() 함수
# 찾고자 하는 문자 또는 문자열의 첫번째 위치(인덱스 값)를 반환함.
# 만일 해당 문자를 찾지 못하면, -1을 반환함.
text = "송아지 송아지 얼룩 송아지"
print(text.find("지"))
print(text.find("얼룩"))
print(text.find("말"))

# 'str.find(sub[, start[, end]])'
# sub: 찾을 문자열
# start: 검색 시작 인덱스
# end: 검색 종료 인덱스(미포함)

# 예제 01
# 찾을 문자열이 여러 번 등장할 땐?
# 송아지는 여러 개 있지만, 항상 첫 번째 위치만 반환.
# find()는 반복 검색용 함수가 아니다.
# 여러 위치를 찾으려면 반복문 또는 정규식을 사용해야 함.
print(f"\n{text.find("송아지")}")

# 예제 02
# start 인자 사용하기
# 로그 분석, 문자열 파싱에서 매우 자주 사용됨.
first = text.find("송아지")
second = text.find("송아지", first + 1)

print(f"\n첫번째 송아지: {first}")
print(f"두번째 송아지: {second}")

# 예제 03
# end 인자를 포함한 범위 검색
# 찾을 인덱스 값은 0, 범위 제한은 인덱스 값 6까지.
fend = text.find("송아지", 0, 6)
print(f"\nend 옵션 사용해 송아지를 찾은 결과: {fend}")

wrend = text.find("송아지", 5, 10)
print(f"검색 범위에 송아지가 없는 경우: {wrend}")

# 예제 04
# if 조건문으로 문자열을 찾으려 할 때 find도 좋긴 한데, in 사용이 더 좋음. (단순 여부 확인 시에만)
# 만일 "if text.find(송아지): print("있음")"이라고 하면 index 0에 있어서 실제로 존재하는데 0을 False로 판단해 없다고 판단하는 오류 발생.
if text.find("송아지") != -1:
    print("\n송아지는 있습니다.")

if "송아지" in text:
    print("in 키워드를 사용해서 찾아도 송아지는 있습니다.")

# 예제 05
# find() vs index()
# 문자열을 찾을 경우 find()와 index() 모두 해당 인덱스 값을 반환.
# 문자열을 찾지 못할 경우 'find(): -1 반환 / index(): 예외 발생.'
# find()는 조건문에 사용 가능하지만, index()는 True/False 등의 요인으로 위해 비추.
# find()는 예외 제어 불필요 / index()는 'try/except'로 예외 제어 필요.

# 결론
# find()는 문자열이 어디에 처음 등장하는지 알려주는 함수. 존재 여부 검사 X. 찾는 문자열의 모든 위치를 반환하지 않음.
```

```python
# count() 함수
# (교재)찾고자 하는 문자 또는 문자열이 포함된 횟수를 반환함.
text = "송아지 송아지 얼룩 송아지"
print(f"교재 예제에서 송아지는 {text.count("송아지")}회 등장.")

# str.count(sub[, start[, end]])
# 문자열 안에 'sub'이 몇 번 등장하는지에 대한 개수를 반환.
# 겹침(overlap)은 세지 않음.
# 못 찾으면 0을 반환.
# 개수를 파악할 때 count()를 쓰자. 존재 여부나 위치를 파악할 때는 각각 'in'과 'find()'를 쓰자.

# 예제 01
# 겹치는 경우 세지 않음.
text_overlap = "aaaa"
print(f"\n겹치는 경우 세지 않아서, 'aa'를 찾는다고 할 때 값은 {text_overlap.count('aa')}회 입니다.")
print(f"'aaa'를 찾는다고 하면 {text_overlap.count('aaa')}회가 반환됩니당.")
```

```python
# startswith() 함수
# 문자열이 특정 문자열로 시작하는지 확인.
# str.startswith(prefix[, start[, end]])
# 문자열이 특정 접두사로 시작하면 'True', 아니면 'False'를 반환.

# 예제 01
filename = "report_2025.csv"
print(f"이 파일은 report로 시작하나요? \u200b/ 답변: '{filename.startswith("report")}'")

# 예제 02
# 튜플을 사용하기.
# 찾는 문자열을 ()로 한 번 더 감싸서 튜플로 만들기.
# 여러 패턴을 검사할 때 좋음.
url = "https://example.com"
# print(f"\nurl은 'http://' 또는 'https://'로 시작하나요? / 답변: '{url.startswith(('http://', 'https://'))}'")
# '/' 앞에 '\u200b'(Zero-width-space)를 입력하면 VSCode에서 링크로 인식하지 않음.
print(f"\nurl은 'http://' 또는 'https://'로 시작하나요? \u200b/ 답변: '{url.startswith(('http://', 'https://'))}'")

# find() 대신 사용하는 이유 01. 의도가 명확함.
# find() 대신 사용하는 이유 02. 버그 가능성이 낮음.
```

```python
# endswith() 함수
# 문자열이 특정 문자열로 끝나는지를 확인하는 함수
# str.endswith(suffix[, start[, end]])
# 문자열이 특정 접미사로 끝나면 True를 반환, 아니면 False를 반환.

# 예제 01
filename = "analysis_result.xlsx"
print(f"파일 확장자가 xlsx 인가요? \u200b/ 답변: {filename.endswith("xlsx")}")

# 예제 02
# 튜플을 활용해 여러 패턴을 검사하기.
image = "photo.jpeg"
print(f"사진 확장자가 '.jpg' 또는 '.png' 또는 '.jpeg' 인가요? \u200b/ 답변: {image.endswith((".jpg", ".png", ".jpeg"))}")

# endswith() 함수는 문자열 분기 처리용 함수.
```

```python
# replace() 함수
# 지정한 문자열을 다른 문자열로 변환하는 함수.
# str.replace(old, new[, count])
# old를 new로 치환함. 기본값은 모두 치환. 그래서 count를 생략 가능.
# 문자열은 불변 객체로, 원본 변경이 안됨.

# 예제 01
text = "송아지"
print(f"(송아지를 강아지로 바꾸기) 송아지 ==> {text.replace("송", "강")}")

# 예제 02
msg = "error error error"
print(f"\n경고를 주의로: {msg} ==> {msg.replace("error", "warning")}")

# 예제 03
# count 인자 사용하기
# count 인자 사용 시 왼쪽부터 지정한 개수만큼만 변환함.
print(f"\nwaring 1개와 error 2개: <{msg}> ==> <{msg.replace("error", "warning", 1)}>")

# 위치에 대한 조건 X
# 패턴에 대한 조건 X
# 무조건 문자열로 치환함.
# 조건부 치환을 하려면 정규식을 사용해야 함.
```

```python
# lower() 함수
# 문자열을 소문자로 변환함.
# 비교 전 전처리에 매우 중요한 함수.

# 예제 01
user_input = "Admin"
low_user_input = user_input.lower()
print(low_user_input)

# 예제 02
# 사용자의 입력을 정규화함.
# 대소문자 버그를 방지함.
role = "ADMIN"
if role.admin == "admin":
    print("관리자")

# lower()은 비교 안정성 확보용 함수.
```

```python
# upper() 함수
# 문자열을 대문자로 변환하는 함수.

# 예제 01
code = "us"
print(code.upper())

# 예제 02
# 코드값 통일
# 데이터의 정합성 유지.
country = "kr"
standard = country.upper()
print(f"나라명은 무조건 대문자를 사용해야 해요. 그래서 kr이 아니라 <{standard}>을 사용해야 합니다.")

# upper()는 표준화용 함수.
```

```python
# 문제 01. 사용자로부터 여러 명의 이름을 입력받아 데이터를 정제하고 분석하는 프로그램 작성하기.
# 요구사항 01. 입력 데이터는 " 김철수, 홍길동, 이영희 "
# 요구사항 02. 데이터 정제는 앞뒤 공백 제거.
# 요구사항 03. 이름을 쉼표와 공백을 기준으로 리스트로 분리하기.
original_data = " 김철수, 홍길동, 이영희 "

refined_data = original_data.strip()
separated_data = refined_data.split(", ")

print(f"=== 사용자 명단 처리 결과 ===")
print(f"원본 데이터: '{original_data}'")
print(f"정제된 데이터: '{refined_data}'")
print(f"분리된 이름 목록: {separated_data}")

for i, name in enumerate(separated_data, 1):                # 인덱스와 iterator의 값이 동시에 필요할 때 'len(range)'보다 enumerate를 쓰는 습관을 들여보자.
    print(f"{i}번째 사용자: {name} (길이: {len(name)}자)")
```
