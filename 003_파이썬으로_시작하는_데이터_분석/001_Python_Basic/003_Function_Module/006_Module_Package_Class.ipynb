```python
# 모듈 생성하기.
# 교재에 나와있는 대로 'diet.py'를 생성함.
```

```python
# 모듈 불러오기 01. import 사용하기.
# 파일을 통째로 불러오기 때문에, 그 안의 모든 메서드를 사용 가능.
# 여러 모듈을 불러와도 구분이 쉬움.
import diet

diet.get_recommend_weight(160, False)
diet.print_valid_menu()
```

```python
# 모듈 불러오기 02. 별명(alias) 사용하기.
# 불러올 모듈명이 길 경우, 사용하면 좋음.
# ex) import numpy as np
import diet as dd

dd.get_recommend_weight(177)
dd.print_valid_menu()
```

```python
# 모듈 불러오기 03. from 사용
# 'from 모듈명 import 모듈 내의 사용할 메서드' 형태로 사용.
# from으로 불러온 모듈에서 사용하고 싶은 메서드만 import 뒤에 작성하면 됨.
# from을 사용하면 모듈명 없이 함수를 직접 호출할 수 있음.
# 모든 함수를 불러오려면 'from 모듈이름 import *' 구문을 사용하면 됨.
from diet import get_recommend_weight, print_valid_menu

get_recommend_weight(177)
print_valid_menu()
```
