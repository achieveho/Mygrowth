```python
import numpy as np
```

```python
# 배열 만들기
# for문은 반복자(iterator)를 사용해 일정한 횟수만큼 반복을 수행.

# 배열 생성 함수
# np.array([]): 초기화 할 값을 지정해 배열을 생성.
# np.zeros(): 값을 0으로 초기화하여 배열을 생성.
# np.ones(): 값을 1로 초기화하여 배열을 생성.
# np.arange(): 수의 순차적인 증감을 이용하여 배열을 생성.
# np.random.rand(): 랜덤한 숫자로 배열을 생성.
```

```python
# np.array()
# 기본적인 배열 생성 함수.
# 'np.array(iterator)' 형태를 사용.

list1 = [1, 2, 3]
arr1 = np.array(list1)
print(arr1, type(arr1))
```

```python
# np.zeros()
# 값을 0으로 초기화 하여 배열을 생성.
# 입력 인자로 shape를 사용해야 함.
# 'np.zeros(shape)' 형태를 사용.

# 0차원 배열(스칼라) 생성
arr0 = np.zeros(shape=(), dtype=int)                                # shape=()을 사용하면 0차원 스칼라 생성.
print(f"<0차원 배열(스칼라)>\n{np.array2string(arr0, separator=", ")}\n{type(arr0)}")

# 1차원 배열 생성하기.
arr1 = np.zeros(5, dtype=int)                                       # shape = 5는 (5, )로서 깅리가 5인 1차원 배열.
print(f"\n<1차원 배열>\n{np.array2string(arr1, separator=", ")}\n{type(arr1)}")

# 2차원 배열 생성하기
arr2 = np.zeros((2, 3), dtype=int)
print(f"\n<2차원 배열>\n{np.array2string(arr2, separator=' ,')}\n{type(arr2)}")   # shape = (2, 3)은 2차원 배열 안에 1차원 배열이 2개 있고, 각 1차원 배열의 길이는 3.

# 3차원 배열 생성하기.
arr3 = np.zeros((3, 3, 3), dtype=int)
print(f"\n<3차원 배열>\n{np.array2string(arr3, separator=', ')}\n{type(arr3)}")

# 배열을 그냥 print()로 출력하면 요소를 공백으로만 구분해버림.
# 그래서 'np.array2string(출력할 배열, separator=', ')'를 사용하면 실제 list처럼 ,로 원소를 구분해서 출력함.
# 그러나 '서버 로그', '모델 학습 로그', '배치 작업 기록'처럼 간결함, 정렬, 일관성이 중요한 분야에서는 공백으로 구분된 NumPy 기본 표현을 사용함.
```

```python
# np.ones()
# 값을 1로 초기화 하여 배열을 생성.
# 'np.ones(shape)' 형태를 사용. 이것도 '.' 보기 싫으면 dtype=int 적고, 출력할 때 array2string을 사용해서 ', '로 구분하게 하자.

# 0차원 배열(스칼라) 생성
arr0 = np.ones(shape=(), dtype=int)
print(f"<0차원 배열(스칼라)>\n{np.array2string(arr0, separator=', ')}\n{type(arr0)}")

# 1차원 배열 생성
arr1 = np.ones(5, dtype=int)        # arr1 = np.ones((5, ), dtype=int)도 같은 의미.
print(f"\n<1차원 배열>\n{np.array2string(arr1, separator=', ')}\n{type(arr1)}")

# 2차원 배열 생성
arr2 = np.ones((3, 5), dtype=int)
print(f"\n<2차원 배열>\n{np.array2string(arr2, separator=', ')}\n{type(arr2)}")

# 3차원 배열 생성
arr3 = np.ones((3, 3, 5), dtype=int)
print(f"\n<3차원 배열>\n{np.array2string(arr3, separator=', ')}\n{type(arr3)}")
```

```python
# np.empty()
# 초기화를 하지 않는 배열 생성 함수.
# 초기화를 하지 않으므로, 의미 없는 쓰레기 값으로 초기화 되어 배열이 생성됨.
# 'np.zeros()', 'np.ones()'보다 배열 생성 속도가 '조금' 빠르나, 쓰레기 값 초기화에 매우 주의해야 함.
aslkdfjas = np.empty(5)
print(f"{np.array2string(aslkdfjas, separator=', ')}")

# 변수명이 겹쳐 그 메모리를 써서 쓰레기 값이 안 생기는 줄 알았는데.. 그게 아닌가 보네. PASS
```

```python
# np.arange() 함수
# python의 range() 함수와 유사하게 연속된 숫자를 채워서 배열을 만듦.
# 'np.arange([시작번호], 끝번호, [건너뛰는 수])
# 끝 번호는 필수 작성, 나머지는 생략 가능.
# 시작번호 defalut: 0, 건너뛰는 수 default: 1
# '끝번호 - 1' 까지 생성.
ex01 = np.arange(10)
print(f"{np.array2string(ex01, separator=', ')}")

# 시작값과 간격을 지정한 ver
ex02 = np.arange(20, 30, 2)
print(f"{np.array2string(ex02, separator=', ')}")
```

```python
# np.random.rand() 함수
# 랜덤으로 값을 생성해 그 값으로 배열을 생성.
# np.random.rand(shape)
example01 = np.random.rand(2, 3)
print(f"{np.array2string(example01, separator=', ')}")

# NumPy의 random 모듈의 주요 함수
# rand(): 균등분포에서 랜덤 추출.
# randint(): 주어진 범위 안에서 랜덤 추출(정수로만)
# randn(): 표준정규분포에서 랜덤 추출(평균 0, 표준편차 1)
# normal(): 정규분포에서 랜덤으로 추출.
# seed(): 랜덤값 생성기의 seed를 지정.
# shuffle(): 리스트나 배열의 순서를 섞음.
```

```python
# reshape() 함수
# 배열의 차원을 바꿈.
# 'array.reshape(shape)'
original = np.arange(16, dtype = int)
print(f"{np.array2string(original, separator=', ')}")

after = original.reshape(4, 4)
print(f"\n{np.array2string(after, separator=', ')}")

aafter = original.reshape(2, 2, 4)
print(f"\n{np.array2string(aafter, separator=', ')}")
```

```python
# 배열의 속성 확인 및 타입 변환
# 배열 특징 01. 같은 종류의 데이터만 담을 수 있음.
# 배열 특징 02. 차원을 가짐. (0차원부터)

# 배열 속성을 확인하는 함수.
# shape(): 배열의 형태 확인.
# dtype(): 배열의 데이터 타입 확인.
# astype(): 배열의 데이터 타입 변경
check = np.arange(16)
print(f"{np.array2string(check, separator=', ')}\n")

check_shape = check.shape       # .shape로 배열의 형태 확인. ()는 쓰지 않음.
print(f"{check_shape}\n")

check_dtype = check.dtype       # .dtype으로 배열의 데이터 타입 확인. ()는 쓰지 않음.
print(f"data type: {check_dtype}\n")

chg_dtype = check.astype(float)
print(f"기존 배열의 데이터 타입: {check_dtype} \u200b/ 변경 후 배열의 데이터 타입: {chg_dtype.dtype}\n")
```

```python
# NumPy 배열의 저장 및 불러오기.
# 작업한 다차원 배열은 Binary 형태로 저장하고, 다시 불러와 사용 가능.
# 컴퓨터 메모리 상에 올라간 데이터 형태 그대로 저장.
# 사람이 읽을 수 있는 형태로 인코딩하거나 데이터 압축 과정 없이 원본 그대로 저장.
# 빠른 속도로 저장 및 불러오기 가능.

# np.save() 함수
# 배열을 저장함.
# np.save(파일이름, 배열이름)
save_arr = np.arange(10).reshape(2, 5)
print(f"{np.array2string(save_arr, separator=', ')}")

np.save('array', save_arr)      # 'array.npy'라는 파일로 저장. (save 함수는 '.npy' 확장자로 파일을 저장.)

# np.load() 함수
# .npy 확장자 파일을 불러옴.
load_arr = np.load('array.npy')         # 파일의 확장자명까지 적어야 함.
print(f"\n{np.array2string(load_arr, separator=', ')}")

# .npy 파일(Binary) 저장의 장점
# 01. 인코딩/디코딩 과정이 없어 속도가 매우 빠름.
# 02. 데이터 타입과 정밀도가 완벽하게 보존됨.
# 03. 압축 없이도 효율적으로 저장 공간을 사용.
# 04. 운영체제나 아키텍처에 관계없이 호환됨.
```

```python
# npz 형식
# 배열의 압축된 형식
# 여러 개의 배열을 하나의 파일로 저장하고, 그 때 저장되는 파일의 확장자는 '.npz'
# 저장하려는 배열을 인자로 지정하여 함수로 호출.
# 'np.savez(파일이름, 저장할_배열): 여러 개의 배열을 하나의 파일로 저장'
# 'np.savez_compressed(파일이름, 저장할_배열): 여러 개의 배열을 압축하여 하나의 파일로 저장.'

# np.savez()
arr001 = np.arange(10)
arr002 = np.arange(20).reshape(4, 5)
arr003 = np.random.rand(3, 3)

saved = np.savez('multiple_arrays', first_array = arr001, second_array = arr002, third_array = arr003)      # 여러 개의 배열을 하나의 파일로 저장. (기존 배열에 새 이름을 부여하여 하나의 파일에 다른 이름의 배열로 각각 저장.)
saved_noname = np.savez('arrays_no_names', arr001, arr002, arr003)                                          # 배열에 이름을 지정하지 않고도 저장 가능.

# np.load()
load_data = np.load('multiple_arrays.npz')

print(f"{load_data.files}")         # 저장된 배열의 이름을 리스트로 출력.

print(f"\n첫번째 배열: {np.array2string(load_data['first_array'], separator=', ')}")     # 개별적으로 배열에 접근.
print(f"\n두번째 배열: {np.array2string(load_data['second_array'], separator=', ')}")    # 변수로 할당 가능. (load_arr01 = load_data['first_array'])처럼 작성하면 됨.
print(f"\n세번째 배열: {np.array2string(load_data['third_array'], separator=', ')}")

load_data.close()       # 사용을 마쳤으면 파일 닫기.

# np.savez_compressed
arr_compress = np.savez_compressed('compressed_arrays', data01 = arr001, data02 = arr002, data03 = arr003)          # 여러 개의 배열을 압축해서 하나의 파일로 저장해, 기존 np.savez() 보다 저장 공간 효율성 증가.
```
