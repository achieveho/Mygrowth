```python
# 문제 06. 종가(Close) 기준의 최댓값, 최솟값, 평균을 각각 구하기.
max_close = df['Close'].max()
min_close = df['Close'].min()
mean_close = df['Close'].mean()

print(f"Close 컬럼의 최댓값: {max_close}")
print(f"\nClose 컬럼의 최솟값: {min_close}")
print(f"\nClose 컬럼의 평균: {mean_close}")
```

```python
# 문제 07. 거래량(Volume) 컬럼의 평균, 중앙값, 표준편차를 각각 구하기.
mean_vol = df['Volume'].mean()
std_vol = df['Volume'].std()
med_vol = df['Volume'].median()

print(f"Volume 컬럼의 평균: {mean_vol}")
print(f"\nVolume 컬럼의 표준편차: {std_vol}")
print(f"\nVolume 컬럼의 중앙값: {med_vol}")
```

```python
# 문제 07. 고가(High)와 저가(Low)의 평균의 차이 구하기.
mean_diff = df['High'].mean() - df['Low'].mean()

print(f"High 컬럼과 Low 컬럼의 평균의 차이: {mean_diff}")
```

```python
# 문제 08. 'Date', 'Close', 'Volume' 컬럼만 선택한 새로운 데이터프레임 생성하기.
df_copy = pd.DataFrame.copy(df)
# df_copy.describe()
df_copy = df.drop(['High', 'Low', 'Open'], axis=1)

print(df_copy)
```

```python
# 문제 09. 최근 10개 데이터만 선택하기.
last_10 = df.tail(10)

print(last_10)
```

```python
# 문제 10. 종가가 백분위수 90% 이상인 날짜가 몇 일인지 출력하기.
quan_90 = df['Close'].quantile(0.9)
dt_quan90 = df['Close'][df['Close'] > quan_90]

print(f"'Close 컬럼의 백분위수 90% 값: {quan_90}")
print(f"\n'Close' 컬럼의 값이 백분위수 90% 이상인 날짜들: \n{dt_quan90}")
```

```python
# 문제 11. 거래량이 평균 거래량보다 많았던 날의 데이터 선택하기.
avg_volume = df['Volume'].mean()
more_avg_volume = df['Volume'][df['Volume'] > avg_volume]


print(f"평균 거래량: {avg_volume}")
print(f"\n거래량이 평균보다 많았던 날짜와 그 값들: \n{more_avg_volume}")
```

```python
# 문제 12. 거래량(Volume) 기준으로 DF를 내림차순 정렬하기.
df_dsc = df.sort_values('Volume', ascending=False)

print(df_dsc)
```

```python
# 문제 13. 'Daily_Change' 컬럼을 생성하고, 종가와 시가의 차이를 계산한 값을 적용하기.
df['Daily_Change'] = df['Close'] - df['Open']

print(df)
```

```python
# 문제 14. 'Change_Rate' 컬럼을 생성하고, 일일 변동률((종가 - 시가) / 시가 * 100)을 계산하기.
df['Change_Rate'] = ((df['Close'] - df['Open']) / df['Open']) * 100

df
```

```python
# 문제 15. 날짜에서 '연도', '월', '요일'을 추출해 각각 'Year', 'Month', 'Weekday'의 새 컬럼으로 생성하기.
# df['Year'] = df['']
# df['Month'] = 
# df['Weekday'] = 
df_cp = df.copy()
df_cp.reset_index(inplace=True)

df_cp['Year'] = df_cp['Date'].dt.year
df_cp['Month'] = df_cp['Date'].dt.month
df_cp['Weekday'] = df_cp['Date'].dt.weekday

df_cp
# print(df_cp)
```

```python
# 문제 16. 월별로 그룹화하여 평균 종가와 총 거래량을 계산하세요.
mean_close = df_cp.groupby('Month').mean(df_cp['Close'])
total_close = df_cp.groupby('Month').sum()

print(f"월별 평균 종가: {mean_close}")
print(f"\n월별 총 거래량")
```

```python
# 문제 17. 연도별로 그룹화하여 최고가와 최저가를 구하기.
highest_price = df_cp.groupby('Year')
lowest_price = df_cp.groupby('Year')
```
