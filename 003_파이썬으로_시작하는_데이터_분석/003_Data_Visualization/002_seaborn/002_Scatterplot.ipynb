```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")
```

```python
# 산점도 (scatterplot)
# 산점도는 x축 변수와 y축 변수의 관계를 확인할 때 사용하는 그래프.
# 각 변수의 값을 x-y 2차원 평면에 점으로 표시함.
# 각각의 점들의 분포 패턴에 따른 상관관계 판단 가능.
# 상관관계의 강도는 점들이 직선에 얼마나 가깝게 분포되어 있는지로 판단함.
# 즉, 점들이 직선에 가까울수록 강한 상관관계를 의미.

# 양의 상관관계: 점들이 오른쪽 위로 상승하는 형태.
# 음의 상관관계: 점들이 오른쪽 아래로 하강하는 형태.
# 무상관관계: 점들이 특별한 패턴 없이 분산된 형태.
```

```python
# scatterplot 기본 문법
# 'sns.scatterplot(x, y)' 형태로 사용.

# parameter 01. x
# x축 변수

# parameter 02. y
# y축 변수

# parameter 03. hue
# grouping을 할 변수 (색상)

# parameter 04. size
# grouping을 할 변수 (점의 크기)

# parameter 05. style
# grouping을 할 변수 (점의 모양)

# parameter 07. data
# DataFrame 또는 다차원 배열
```

```python
# tips dataset을 scatterplot을 사용해 상관관계 실습하기.
# 01. 음식 가격과 팁 간 상관관계 구하기.
sns.scatterplot(x=df['total_bill'], y=df['tip'])

# max 값이었던 10달러 tip은 딱 1건 존재.
# 음식 가격이 커질수록, 지불하는 팁의 금액도 커짐을 알 수 있음.
# 6달러 미만의 팁이 과반수.
# 근데 뭔가 드라마틱하게(엄청 가파르게) 양의 상관관계가 있다고 하기에는 좀 애매한 것 같은데..
```

```python
# 02. 음식 가격과 팁 금액을 성별로 구분하여 상관관계를 확인하기.
# hue(색상) parameter를 사용하기.
sns.scatterplot(x=df['total_bill'], y=df['tip'], hue=df['sex'])

# 남성이 여성보다 팁 지불 금액이 더 높음.
# 여성의 최고 팁 지불 금액은 6~7달러 사이, 남성의 최대 팁 지불 금액은 10달러(전체 중 최댓값)임을 확인할 수 있음.
# 팁 지불 금액 최솟값은 여성인 것으로 확인됨.
```

```python
# 03. 크기를 활용한 3차원 정보 표현 (Bubble chart)
# size parameter를 사용해 인원 수에 따른 값의 분포 확인하기.
sns.scatterplot(x=df['total_bill'], y=df['tip'], hue=df['sex'], size=df['size'])

# 인원 수에 따라 점의 크기가 달라짐.
# 근데 이 구문으로 나온 결과물을 보면, 각 점이 정확히 몇 명인지 구별하기가 힘듦.
# Bubble chart는 값의 차이가 클 때 사용하는 것이 효과적이라고 하네.
```

```python
# 04. 다차원 시각화
# 산점도는 가로축, 세로축 이외에도 점의 색상, 크기 등을 사용해 3가지 이상의 변수의 상관관계 표시 가능.
# 2차원 평면에 시각화를 하지만, 2차원 이상의 값을 표현할 수 있다는 장점이 있음.
sns.scatterplot(x=df['total_bill'], y=df['tip'], hue=df['sex'], size=df['size'], style=df['time'], alpha=0.7)

# 성별을 색상으로 구분하고,
# 인원 수를 점의 크기로 구분하고,
# 점심과 저녁을 점의 모양으로 구분했음.
# alpha는 투명도를 조절하는 parameter.
# 저녁 식사를 한 남성 고객들의 tip 지불 금액이 동시간에 방문한 여성 고객들보다 더 높음.
```
