```python
import seaborn as sns
import numpy as np
import pandas as pd

df = sns.load_dataset('tips')
```

```python
# Regplot
# 회귀선
# 회귀: 한 바퀴 돌아서 본디의 자리나 상태로 돌아오는 것. (사전적 의미)
# Regplot은 개별 데이터 분포를 대표할 수 있는 하나의 선인 회귀(Regression)선을 표시하는 그래프.
# 산점도에 표시된 모든 데이터 포인트들의 전체적인 경향을 하나의 직선으로 요약해서 보여줌.
```

```python
# Regplot 기본 문법
# 'sns.regplot(x, y)'

# parameter 01. x
# x축 변수
# parameter 02. y
# y축 변수
# parameter 03. scatter
# 산점도를 같이 표시할지에 대한 여부
# parameter 04. ci
# 신뢰구간 (기본값: 95%)
# parameter 05. data
# Dataframe 또는 다차원 배열.
```

```python
# 기본 회귀선 그래프 생성하기.
sns.regplot(x=df['total_bill'], y=df['tip'])

# 'scatter'의 기본값은 'True'로, 인자값 조정을 생략하면 무조건 산점도도 같이 표시됨.
# 회귀선이 중간에 표시되어, 데이터의 전체적인 경향을 나타냄.
```

```python
# 산점도 없이 회귀선만 표시하기.
sns.regplot(x=df['total_bill'], y=df['tip'], scatter=False)

# 산점도를 표시하지 않으니까, tip(y축)의 표기 단위 및 min과 max 값이 달라짐.
# 회귀선의 기울기도 좀 달라짐.
# 단순한 전체적인 경향을 파악하는 용도로는 괜찮을 듯.
```

```python
# 신뢰구간 이해하기.
# 회귀선 주변에 회귀선과 동일한 색상의 연하게 칠해진 영역이, 신뢰구간임.
# 신뢰구간은 기존 데이터 분포가 회귀선을 기준으로 얼마나 퍼져 있는지 확인하는데 도움이 됨.
# 기본값은 95%.
# 실제 paramter 조정 시 정수값으로 '%' 기호 없이 입력하면 되고, 신뢰구간을 표시하지 않으려면 'None'으로 설정하면 됨. 
sns.regplot(x=df['total_bill'], y=df['tip'], scatter=False, ci=None)

# 신뢰구간도 표시를 안 하니까, 그냥 선 그래프처럼 나오네.
```

```python
# 회귀선은 데이터의 상관관계 파악 및 데이터의 경향성을 확인해 새 데이터가 들어왔을 때 예측하는 것에도 도움이 됨.
# 새 고객이 방문했을 때 총 지불한 음식이 30달러라면, 약 4달러 정도의 팁을 지불할 것이라고 예측할 수 있다는 의미.

# 이전까지 한 번도 발생하지 않았던 일에 대해서도 예측 가능.
# 음식 가격으로 100달러를 지불했을 때는, 50달러에 6달러로 균일하게 양의 상관관계를 보이기 때문에 12달러를 팁으로 지불할 것이라고 예측 가능하다는 의미.

# Regplot은 두 변수 간의 전체적인 경향을 파악하고자 할 때 선택하는 것을 권장.
# 특히 미래 값 예측, 데이터의 선형 관계를 정량적으로 확인하고자 할 때 적합함.
# 신뢰구간을 통해 예측의 불확실성을 표현하고자 할 때에도 유용.

# Scatterplot은 개별 데이터의 포인트 분포 패턴을 자세히 관찰하고자 할 때 선택하는 것을 추천.
# 이상치 식별에 효과적, 여러 그룹을 색상이나 크기로 구분해 표시하고자 할 때 유용.
```
