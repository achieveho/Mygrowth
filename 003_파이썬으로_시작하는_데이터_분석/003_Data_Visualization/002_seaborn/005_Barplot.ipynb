```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df = sns.load_dataset("tips")
```

```python
# 막대그래프 - Barplot, Countplot 둘 다 사용.
# 수치를 막대의 길이로 표현해 시각화하는 방법.
# 값들의 차이가 미미하거나 표시할 막대의 수가 많은 경우에 그래프를 이해하기 어려운 문제 존재.
# 그럴 때는 선그래프를 사용하거나 막대의 색상을 조절해 시각화하는 것이 좋음.
```

```python
# barplot 기본 문법
# 'sns.barplot(x, y)'

# parameter 01. x
# x축 변수

# parameter 02. y
# y축 변수

# parameter 03. hue
# 색상으로 grouping 할 변수.

# parameter 04. orient
# 막대의 방향
# vertical: 기본 / horizontal: 수평

# parameter 05. errorbar(원래는 ci인데, FutureWarning 때문에 앞으로 ci 대신 errorbar를 쓸 것임.)
# 신뢰구간의 값을 설정. (0 ~ 100의 정수)
# 기본값 - 95%
# 신뢰구간을 표시하지 않을 경우: 0으로 설정.

# parameter 06. data
# Dataframe 또는 다차원 배열.
```

```python
# 인원수에 따른 tip 지불 금액을 파악하기 위한 막대그래프 생성하기.
sns.barplot(x=df['size'], y=df['tip'])

# 각 막대그래프의 중심에 있는 직선은 신뢰구간.
```

```python
# 신뢰구간을 표시하지 않는 막대그래프 생성하기.
sns.barplot(x=df['size'], y=df['tip'], errorbar=('ci', 0))
```

```python
# 만일 막대의 길이가 비슷해 차이가 미미할 땐, 다른 형태의 그래프를 그려 확인하는 것이 더 좋음.
# 정답은 없고, 데이터의 특징 및 분포에 따라 적절한 그래프를 선택해 생성할 수 있도록 다양한 방법을 숙지하는 것이 좋음.
sns.barplot(x=df['day'], y=df['tip'])

# 이게 좋지 않은 그래프네.
# 요일 별 tip의 차이가 별로 나지 않는다는 것을 알 수는 있네.
```

```python
# countplot 사용하기.
# countplot이란, barplot과 모양은 유사하지만 y축의 값을 x축 변수의 개수로 표시하는 그래프.
# countplot은 y축에 변수의 개수를 표시하기 때문에, x축에 설정할 변수만 지정함.
sns.countplot(x=df['day'])

# 요일별로 지불한 tip의 금액은 알 수 없음.
# 단, 요일별 고객 방문 횟수를 알 수 있음.
# 금요일에 팁 지불 총액 및 total_bill의 값이 제일 적을 것이라고 유추해봄.
```

```python
# countplot은 'value_counts()' method의 결과를 막대그래프로 시각화 한 것.
df['day'].value_counts()
```

```python
# 시각화를 사용해 데이터를 탐색하면, 숫자나 표보다 훨씬 더 데이터를 이해하기 쉽고 기억하기 쉬움.

# 막대그래프는 가장 직관적이고 이해하기 쉬운 시각화 방법 중 하나.
# 프레젠테이션이나 보고서에서 핵심 메세지를 전달할 때 유용.
# 막대그래프를 사용할 때 막대 순서는 크기나 시간 등 의미있는 순서로 정렬하는 것이 효과적임.
# 색상을 활용해 중요한 카테고리를 강조하고, 축 제목과 범례를 명확히 작성해야 함.
# 막대 간격은 너무 좁거나 넓지 않게 적절하게 하는 것이 중요함.

# 막대그래프는 카테고리별 수치를 비교할 때 적합.
# 그룹 간 평균값 비교, 빈도 또는 개수를 표시할 때 유용.
# 순위나 크기 순서를 강조할 때도 유용.

# 막대의 길이가 비슷할 때는 선그래프나 점그래프가 더 좋음.
# 시간 변화에 따른 추이를 볼 때도 선그래프가 더 좋음.
# 연속형 변수 간 관계를 파악할 땐 산점도.
# 분포를 확인하고자 할 때는 히스토그램이나 박스플롯이 좋음.
```

```python
# pointplot
# 막대 그래프의 변형된 version
# 막대그래프에서 막대를 제외하고, 데이터의 평균값을 선으로 잇고 신뢰구간을 함께 표시한 그래프.
sns.pointplot(x=df['day'], y=df['tip'])

# 요일별 세로 선이 요일별 신뢰구간.
# 점이 x축 변수의 평균값.
```
