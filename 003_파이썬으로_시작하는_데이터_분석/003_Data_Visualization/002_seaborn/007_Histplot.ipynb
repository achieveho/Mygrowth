```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df = sns.load_dataset("tips")
```

```python
# 히스토그램 - histplot
# histplot이란, 변수의 분포를 막대형 그래프를 사용해 표시하는 방법
# x축에 계급, y축에 도수를 나타낸 뒤, 각 계급의 크기를 가로의 길이로, 도수를 세로의 길이로 하는 직사각형을 차례로 그려 표현함.

# 히스토그램은 데이터의 분포를 이해하는 가장 기본적이면서 강력한 도구.
# 데이터의 정규성 확인, 이상치 탐지, 변수 변환의 필요성 판단 등 탐색적 데이터 분석에서 필수적으로 사용됨.

# 막대그래프와 히스토그램의 차이점
# 히스토그램은 가로축과 세로축이 연속적이며, 막대들 사이에 간격이 없음.
```

```python
# histplot 기본 문법
# 'sns.histplot(data)'

# parameter 01. data
# Dataframe 또는 다차원 배열

# parameter 02. hue
# 색상으로 grouping 할 변수.

# parameter 03. bins
# 막대의 개수

# parameter 04. kde
# 막대를 부드럽게 연결한 선을 표시할지에 대한 여부.
# Kernel Destiny Estimation의 약자. 커널 밀도 추정.
```

```python
# 'total_bill'의 분포를 확인하는 기본 histoplot 그려보기.
sns.histplot(data=df['total_bill'])

# barplot과 달리, 모든 막대가 붙어있음.
# 이 histplot으로 음식값이 낮은 경우가 높은 경우보다 훨씬 많다는 정보 확인 가능.
# histplot의 막대의 높이는 빈도, 즉 도수의 크기를 나타냄.
```

```python
# bins parameter로 막대의 개수 조정하기.
sns.histplot(data=df['total_bill'], bins=5)

# 5개의 막대가 생성됨.
# 막대의 개수를 줄여도 음식값의 분포를 확인할 수는 있음.
# 그러나, 조심해야 함.

# 막대의 개수가 너무 적으면: 분포의 세부사항을 놓칠 수 있음.
# 막대의 개수가 너무 많으면: 노이즈가 많아져 패턴 파악이 어려움.
# 일반적인 가이드: '데이터 개수의 제곱근' 또는 'bins="struges"'로 사용.
```

```python
# kde parameter를 활용하여 histplot 생성하기.
# KDE란, 데이터의 밀도를 표시한 값.
# 히스토그램에서 확인할 수 있는 데이터 분포를 부드럽게 연결한 선. (곡선으로)
sns.histplot(data=df['total_bill'], kde=True)

# 일반적인 데이터는 종 모양의 정규분포인 경우가 많아, 평균값을 중심으로 양 옆으로 점진적인 감소 모양을 보이는 히스토그램이 많은 편.
# 'total_bill'은 kde 곡선이 왼쪽으로 치우쳐저 있어, 데이터가 전체 범위에서 수치가 낮은 쪽에 몰려있음을 확인할 수 있음.
```

```python
# kde 곡선의 모양 해석

# 정규 분포: 종 모양, 평균을 중심으로 대칭.
# 왼쪽 치우침(right skewed): 긴 꼬리가 오른쪽에 위치함.
# 오른쪽 치우침(left skewed): 긴 꼬리가 왼쪽에 위치함.
# 이봉분포(bimodal): 두 개의 봉우리
# 균등분포: 평평한 형태
```

```python
# 분포 시각화를 위한 다른 방법들
# kdeplot과 rugplot 사용 가능.
```

```python
# kdeplot
# 막대 없이 kde 곡선으로만 나타낸 그래프.
# x축은 계급이고, y축은 밀도를 나타내는 'destiny'로 표시됨.
sns.kdeplot(data=df['total_bill'])

# 10달러 ~ 20달러 사이의 음식값에서 밀도가 제일 높은 것이 확인됨.
# 즉, 데이터의 분포가 10달러 ~ 20달러 사이에서 제일 많다는 것을 의미.
```

```python
# rugplot
# 작은 실 모양의 선이 데이터의 분포에 따라 표시되는 그래프.
sns.rugplot(data=df['total_bill'])

# 당장의 total_bill column만 보면, rugplot보다는 kdeplot이 더 가시성이 좋다고 느껴짐.
```

```python
# histplot vs countplot
# histplot은 연속형 변수의 분포를 확인할 때 사용하는 그래프.
# 범주형 데이터(categorical)를 사용해 그래프를 그리면 countplot과 유사한 형태로 생성됨.

# 범주형 데이터 히스토그램
sns.histplot(data=df['day'])
```

```python
# 범주형 데이터 countplot
sns.countplot(x=df['day'])
```
