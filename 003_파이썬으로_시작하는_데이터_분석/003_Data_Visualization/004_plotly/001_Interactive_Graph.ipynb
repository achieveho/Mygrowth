```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
# matplotlib과 seaborn은 정적 그래프로 데이터를 시각화 함.
# plotly는 interactive한 웹 기반 데이터 시각화를 위한 오픈소스 라이브러리.
```

```python
# 'plotly.graph_objects'는 모든 그래프 요소를 세밀하게 제어할 수 있는 저수준 인터페이스를 제공함.
# 'plotly.express'는 간단한 함수 호출만으로 복잡한 시각화를 생성할 수 있는 고수준 인터페이스.
```

```python
# Interactive Visualize는 사용자가 그래프와 직접 상호작용 할 수 있는 시각화 방식.
# 전통적인 정적 그래프와 달리 확대, 축소, 필터링, 호버 등의 기능을 제공함.

# Interactive Visualize의 장점 01. 사용자가 데이터를 능동적으로 탐색할 수 있음.
# 마우스를 올리면 상세 정보가 표시되고,
# 특정 영역을 확대해 세부 사항 확인 가능.

# Interactive Visualize의 장점 02. 하나의 그래프로 다차원 데이터를 효과적으로 표현 가능.
# 색상, 크기, 애니메이션 등을 활용해 여러 변수를 동시에 시각화 가능.

# Interactive Visualize의 장점 03. 웹 브라우저에서 바로 실행되어 별도의 SW 설치 없이 공유 가능.
# 분석 결과를 HTML 파일로 저장하면 누구나 웹 브라우저에서 확인 가능.
# 시계열 데이터 시각화 시, 사용자가 직접 기간을 선택하고 특정 시점의 값 확인 가능.
```

```python
# plotly 맛보기
data = {'x': [1, 2, 3, 4], 'y': [10, 11, 12, 13]}

fig = px.line(data, x='x', y='y', title='첫 번째 Plotly 그래프')
fig.show()

# 확대, 축소 가능.
# 마우스로 영역 이동도 가능한 듯.
# 전체 화면도 되나?
# PNG 파일로 그래프 다운로드 가능.
```
