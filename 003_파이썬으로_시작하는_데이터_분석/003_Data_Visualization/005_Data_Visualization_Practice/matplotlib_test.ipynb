```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager as fm

print("active matplotlibrc :", matplotlib.matplotlib_fname())
print("font.family         :", plt.rcParams["font.family"])
print("font.sans-serif     :", plt.rcParams["font.sans-serif"][:5])
print("axes.unicode_minus  :", plt.rcParams["axes.unicode_minus"])
print("figure.figsize      :", plt.rcParams["figure.figsize"])
print("figure.dpi          :", plt.rcParams["figure.dpi"])
print("savefig.dpi         :", plt.rcParams["savefig.dpi"])
print("savefig.bbox        :", plt.rcParams["savefig.bbox"])
print("Noto Sans KR found  :", any(f.name == "Noto Sans KR" for f in fm.fontManager.ttflist))

x = np.linspace(-5, 5, 400)
y = x**3 - 3*x

fig, ax = plt.subplots()
ax.plot(x, y, label="y = x³ - 3x")
ax.set_title("Matplotlib 한글 폰트 / 음수 부호 / 저장 테스트")
ax.set_xlabel("가로축")
ax.set_ylabel("세로축")
ax.legend()

plt.show()

fig.savefig("matplotlib_korean_test.png")
print("saved -> matplotlib_korean_test.png")
```
